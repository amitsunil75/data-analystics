To find the number of clusters using a dendrogram, you typically look for the "elbow" or "knee" point where the dendrogram has the most significant change in the lengths of its branches. This point indicates the optimal number of clusters.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage

# Generate synthetic data
X = np.random.randn(100, 2)

# Perform hierarchical clustering
Z = linkage(X, method='ward')

# Plot the dendrogram
plt.figure(figsize=(10, 5))
dendrogram(Z)
plt.title('Hierarchical Clustering Dendrogram')
plt.xlabel('Sample Index')
plt.ylabel('Distance')
plt.show()
```
